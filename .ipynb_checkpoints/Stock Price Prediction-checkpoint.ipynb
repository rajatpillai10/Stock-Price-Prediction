```python
import pandas as pd
import numpy as np
```

```python
x_train= pd.read_csv("x_train.csv").values
```

```python
pd.read_csv("x_test.csv")
```

```python
y_train=pd.read_csv("y_train.csv").values
```

```python
from sklearn.preprocessing import MinMaxScaler
sc = MinMaxScaler(feature_range = (0, 1))
x_train = sc.fit_transform(x_train)
y_train = sc.transform(y_train)
```

```python
x=[]
y=[]
for i in range(40,1257):
    x.append(x_train[i-40:i, 0])
    y.append(y_train[i,0])
```

```python
x_train,y_train=np.array(x),np.array(y)
```

```python
x_train = np.reshape(x_train, (x_train.shape[0], x_train.shape[1], 1))
```

```python
from keras.models import Sequential
from keras.layers import *
model=Sequential()
model.add(LSTM(32,input_shape=(None,1),return_sequences=True))
model.add(Dropout(0.5))
model.add(LSTM(32,return_sequences=True))
model.add(Dropout(0.5))
model.add(LSTM(32))
model.add(Dropout(0.5))
model.add(Dense(units=1))
model.compile(optimizer='adam',loss='mean_squared_error',metrics=['acc'])
model.summary()
```

```python
from keras.callbacks import ModelCheckpoint
modelcheckpoint = ModelCheckpoint("best_model.h5",save_best_only=True)
hist = model.fit(x_train,y_train, epochs = 50, batch_size = 32, validation_split=0.2,callbacks=[modelcheckpoint])
```

```python
model.load_weights("best_model.h5")
```

```python
X_Train=pd.read_csv("x_train.csv")
X_Test=pd.read_csv("x_test.csv")
data = pd.concat((X_Train,X_Test), axis=0)
```

```python
inputs = data[len(data) - 19 - 40:].values
inputs = inputs.reshape(-1,1)
inputs  = sc.transform(inputs)
```

```python
x_test=[]
for i in range(40,inputs.shape[0]):
    x_test.append(inputs[i-40:i,0])
x_test = np.array(x_test)
```

```python
x_test=x_test.reshape((x_test.shape[0],x_test.shape[1],1))
pred=model.predict(x_test)
pred=sc.inverse_transform(pred)
```

```python
pred=np.round(pred,2)
pred
```

```python
result = pd.DataFrame(pred, columns=['High'])
```

```python
result.to_csv("prediction.csv", index=False)
```
